```python
#buat function untuk mengekstrak data dari file JSON
def ekstrakDataJSON(namafile):
  df = pd.read_json(namafile)
  return  df
```

```python
Columns=['Name','Market Cap (US$ Billion)']
def ekstrak():
  #mendefind variabel dulu
  data = pd.DataFrame(columns=Columns)

  #lakukan proses ekstrack
  for filejson in glob.glob("bank_market_cap_1.json"):
    data = data.append(ekstrakDataJSON(filejson),ignore_index=False)
  
  return data
```

```python
#mengambil data dari file CSV untuk exchange rate GBP
df = pd.read_csv('exchange_rates.csv',index_col=0)
exchange_rate = df.loc['GBP','Rates']

#transfrom data
def transform(data):
  data['Market Cap (US$ Billion)'] = data['Market Cap (US$ Billion)'] * exchange_rate
  data['Market Cap (US$ Billion)'] = data['Market Cap (US$ Billion)'].round(3)
  data = data.rename(columns={"Market Cap (US$ Billion)": "Market Cap (GBP$ Billion)"})
  return data
```

```python
#load data
def load(data, filename):
    data.to_csv(filename, index=False)
```

```python
#eksekusi func ekstrak
eks = ekstrak()
eks.head(3)
```

```python
#eksekusi func Transform
trans = transform(eks)
trans.head()
```

```python
#load
load(trans,"Baru.csv")
```
